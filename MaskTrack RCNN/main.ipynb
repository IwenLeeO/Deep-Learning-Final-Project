```python
from __future__ import division

import mmcv
from mmcv import Config
from mmcv.runner import load_checkpoint
from mmdet.models import build_detector
from mmdet import __version__
from mmdet.datasets import get_dataset
from mmdet.apis import (train_detector, init_dist, get_root_logger,
                        set_random_seed, inference_detector, show_result)
from mmdet.models import build_detector
import torch

cfg = mmcv.Config.fromfile('configs/masktrack_rcnn_x101_fpn_1x_cellcount.py')
```

```python
#####   Train   ######
model = build_detector(cfg.model, train_cfg=cfg.train_cfg, test_cfg=cfg.test_cfg)
train_dataset = get_dataset(cfg.data.train)
logger = get_root_logger(cfg.log_level)
logger.info('Distributed training: {}'.format(False))

train_detector(
    model,
    train_dataset,
    cfg,
    distributed=False,
    validate=True,
    logger=logger)
```

```python
#####   Predict   ######
_ = load_checkpoint(model, 'work_dirs/masktrack_rcnn_x101_fpn_1x_cellcount/epoch_50.pth')
```
